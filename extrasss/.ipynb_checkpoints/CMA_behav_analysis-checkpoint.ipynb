```python
AB1test = []
AB2test = []
AB3test = []
BCXYtest = []
BCXY2test = []
BCXY3test = []
ACtest = []
for a in csv_list:
    if 'AB1test' in a:
        AB1test.append(a)
    if 'AB2test' in a:
        AB2test.append(a)
    if 'AB3test' in a:
        AB3test.append(a)
    if 'BCXYtest' in a:
        BCXYtest.append(a)
    if 'BCXY2test' in a:
        BCXY2test.append(a)
    if 'BCXY3test' in a:
        BCXY3test.append(a)
    if 'ACtest' in a:
        ACtest.append(a)
```

```python
ne_t_ACrt = []
ne_n_ACrt = []

fe_t_ACrt = []
fe_n_ACrt = []

for b in ACtest:
    one_file = pd.read_csv(b, header = 0)
    one_file.columns = one_file.columns.str.replace(".", "_")
    ACresp = one_file.key_resp_keys
    trial_ACrt = one_file.key_resp_rt
    objectcue = one_file.ACobjectA
    AB1faces = one_file.AB1faces
    AB1objectsA = one_file.AB1objectsA
    chk_empty = one_file.test1_thisTrialN
    ne_pt_ACrt= []
    fe_pt_ACrt = []
    fe_trials = []
    ne_trials = []
    for index, row in one_file.iterrows():
        if np.isnan(chk_empty[index]) == False and AB1faces[index][-8:-6] == "NE":
            ne_trials.append(AB1objectsA[index])
        if np.isnan(chk_empty[index]) == False and AB1faces[index][-8:-6] == "FE":
            fe_trials.append(AB1objectsA[index])

    
    for index, row in one_file.iterrows():
        if np.isnan(ACresp[index]) == False and objectcue[index] in ne_trials:
            ne_pt_ACrt.append(trial_ACrt[index])
        if np.isnan(ACresp[index]) == False and objectcue[index] in fe_trials:
            fe_pt_ACrt.append(trial_ACrt[index])
    ne_t_ACrt.append(np.mean(ne_pt_ACrt))
    ne_n_ACrt.append(len(ne_pt_ACrt))
    fe_t_ACrt.append(np.mean(fe_pt_ACrt))
    fe_n_ACrt.append(len(fe_pt_ACrt))
```
